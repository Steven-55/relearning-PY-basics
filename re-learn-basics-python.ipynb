```python
# Question 1: Check if a number is even
def is_even(x):
    return x % 2 == 0

# Question 2: Reverse a string
def reverse_string(x):
    return x[::-1]

# Question 3: Find the maximum number in a list
def find_max(x):
    a = float('-inf')
    for n in x:
        if a < n:
            a = n
    return a

# Question 4: Calculate the factorial of a number
def factorial(x):
    fact = 1
    if x == 0:
        return 1
    for n in range(1, x + 1):
        fact *= n
    return fact

# Question 5: Generate the nth Fibonacci number
def fibonacci(x):
    if x == 1:
        return 1
    if x == 0:
        return 0
    return fibonacci(x - 1) + fibonacci(x - 2)

# Question 6: Check if a string is a palindrome
import re

def is_palindrome(x):
    x = re.sub(r'[^A-Za-z0-9]', '', x)
    return x.lower() == x[::-1].lower()

# Question 7: Check if two strings are anagrams
import re

def is_anagram(x, y):
    x = re.sub(r'[^A-Za-z0-9]', '', x)
    y = re.sub(r'[^A-Za-z0-9]', '', y)
    return sorted(x.lower()) == sorted(y.lower())

# Question 8: Check if a number is prime
import math

def is_prime(x):
    if x <= 1:
        return False
    for n in range(2, int(math.sqrt(x)) + 1):
        if x % n == 0:
            return False
    return True

# Question 9: Check if a number is an Armstrong number
def is_armstrong(x):
    if x < 0:
        return False
    num = str(x)
    total = sum(int(digit) ** len(num) for digit in num)
    return total == x

# Question 10: Count the number of vowels in a string
def count_vowels(x):
    vowels = {'a', 'e', 'i', 'o', 'u'}
    return sum(1 for char in x if char.lower() in vowels)

def second_largest(x):
    unique = list(set(x))
    if len(unique)<=2:
        return None
    unique.sort()
    return unique[-2]
print(second_largest([10, 20, 4, 45, 99]))  # Output: 45
print(second_largest([1, 1, 1, 1]))         # Output: None
print(second_largest([2]))                 # Output: None

import re
def count_unique_words(x):
    x=re.sub(r'[^A-Za-z]'," ",x)
    x = x.lower().split()
    x = list(set(x))
    return len(x)
print(count_unique_words("Hello world! Hello again."))  # Output: 3
print(count_unique_words("Python is fun. Python is awesome."))  # Output: 4
print(count_unique_words(""))  # Output: 0

print('====================================')
def longest_word(x):
    x = re.sub(r'[^A-Za-z\s]', '', x) 
    words = x.split()
    if not words:  
        return ""
    return max(words, key=len) 

print(longest_word("The quick brown fox jumped over the lazy dog."))  # Output: "jumped"
print(longest_word("Python programming is fun!"))                    # Output: "programming"
print(longest_word(""))  # Output: ""
```
